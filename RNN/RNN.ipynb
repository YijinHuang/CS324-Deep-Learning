```python
import numpy as np
import torchvision
from train import train
import matplotlib.pyplot as plt

# default constants
INPUT_LENGTH = 10
INPUT_DIM = 10
NUM_CLASSES = 10
NUM_HIDDEN = 32
OUTPUT_DIM = 10
NUM_BATCH_SIZE = 128
LEARNING_RATE = 0.001
TRAIN_STEPS = 10000
MAX_NORM = 10.0
```

```python
# Train on T=10
print('Train on T = 10:')
epochs, accs, losses, acc_10 = train(10, print_log=True)
```

```python
# accuracy curve of validataion data
acc_plot = plt.plot(epochs, accs)
plt.ylim([0, 1.1])
plt.show()
plt.close()
```

```python
# Test length
_, _, _, acc_3 = train(3, print_log=False)
print('T=3, acc={}'.format(acc_3))
_, _, _, acc_5 = train(5, print_log=False)
print('T=5, acc={}'.format(acc_5))
_, _, _, acc_15 = train(15, print_log=False)
print('T=15, acc={}'.format(acc_15))
_, _, _, acc_20 = train(20, print_log=False)
print('T=15, acc={}'.format(acc_20))
_, _, _, acc_25 = train(25, print_log=False)
print('T=25, acc={}'.format(acc_25))
```
